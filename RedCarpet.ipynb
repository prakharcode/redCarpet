```python
from fuzzywuzzy import fuzz
```

```python
sim_lis=[]
def chng(x):
    for i in df_primary['Title']:
        if fuzz.ratio(i,x)>=80.0:
            sim_lis.append((i,x))
            return i
    return np.nan
df_movies['Title']= df_movies['Title'].apply(chng)
```
